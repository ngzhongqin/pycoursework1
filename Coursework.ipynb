```python
import numpy as np
import matplotlib.pyplot as plt

#read data from a file
title = "Resale prices by flat type"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()

data = np.genfromtxt('resale-flat-prices-based-on-registration-date-from-jan-2015-onwards.csv', 
                     skip_header=1,
                     dtype=[('month','U30'),
                            ('town','U30'),
                            ('flat_type', 'U30'),
                            ('block', 'U30'),
                            ('street_name','U50'),
                            ('storey_range','U10'),
                            ('floor_area_sqm','i8'),
                            ('flat model','U50'),
                            ('lease_commence_date','U10'),
                            ('remaining_lease','i8'),
                            ('resale_price','i8')],
                     delimiter= "," ,missing_values=['na','-'],
                     filling_values=[0])

labels = list(set(data['flat_type']))
labels.sort()
print(labels)
levels = np.arange(0,len(labels))
flat_price = data[['flat_type','resale_price']]

values = flat_price['resale_price']

price_1room = values[flat_price['flat_type'] == '1 ROOM']
price_2room = values[flat_price['flat_type'] == '2 ROOM']
price_3room = values[flat_price['flat_type'] == '3 ROOM']
price_4room = values[flat_price['flat_type'] == '4 ROOM']
price_5room = values[flat_price['flat_type'] == '5 ROOM']
price_Exec = values[flat_price['flat_type'] == 'EXECUTIVE']
price_Multigen = values[flat_price['flat_type'] == 'MULTI-GENERATION']

values_combined = [price_1room, price_2room, price_3room, price_4room, price_5room, price_Exec, price_Multigen]
print(len(values_combined))

plt.figure(2, figsize=(20,10))
plt.title(title,fontsize=25, fontweight="bold")
plt.ylabel('Resale price',fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16,rotation='vertical')
bp_dict = plt.boxplot(values_combined, labels=labels, patch_artist=True)


## change outline color, fill color and linewidth of the boxes
for box in bp_dict['boxes']:
    # change outline color
    box.set( color='orchid', linewidth=2)
    # change fill color
    box.set( facecolor = 'plum' )
    
## change color and linewidth of the whiskers
for whisker in bp_dict['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp_dict['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp_dict['medians']:
    median.set(color='indigo', linewidth=2)

## change the style of fliers and their fill
for flier in bp_dict['fliers']:
    flier.set(marker='D', color='#e7298a', alpha=0.3)

print(bp_dict.keys())

for line in bp_dict['medians']:
    # get position data for median line
    x, y = line.get_xydata()[1] # top of median line
    # overlay median value
    plt.text(x, y, '%.1f' % y,
         horizontalalignment='center',fontsize=14) # draw above, centered

fliers = []
for line in bp_dict['fliers']:
    ndarray = line.get_xydata()
    if (len(ndarray)>0):
       max_flier = ndarray[:,1].max()
       max_flier_index = ndarray[:,1].argmax()
       x = ndarray[max_flier_index,0]
       print("Flier: " + str(x) + "," + str(max_flier))
     
       plt.text(x,max_flier,'%.1f' % max_flier,horizontalalignment='center',fontsize=14,color='darkslateblue') 

plt.show()
```
